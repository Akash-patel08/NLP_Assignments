# Assignment 5 :- Semantic Relationship Identification using WordNet

Objective: Identify synonymy, antonymy, and hypernymy using WordNet.

```python
import nltk
from nltk.corpus import wordnet as wn
```

```python
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
text = 'The dog is happy and good'
words = text.split()
print(words)
```

```python
def get_synonyms(word):
    return {l.name() for s in wn.synsets(word) for l in s.lemmas()}
```

```python
def get_antonyms(word):
    return {l.antonyms()[0].name() for s in wn.synsets(word)
            for l in s.lemmas() if l.antonyms()}
```

```python
def get_hypernyms(word):
    return {h.name() for s in wn.synsets(word) for h in s.hypernyms()}
```

```python
for word in words:
    print('\nWord:', word)
    print('Synonyms:', get_synonyms(word))
    print('Antonyms:', get_antonyms(word))
    print('Hypernyms:', get_hypernyms(word))
```
